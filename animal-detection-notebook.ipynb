```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define constants
train_dir = '/kaggle/input/animals-detection-images-dataset/test'
test_dir = '/kaggle/input/animals-detection-images-dataset/test'
img_size = 48  # Size of images (assuming square images)
batch_size = 32
num_classes = 80  # Number of classes (animals)

# Data Augmentation for Training (Example)
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

# Validation Data Generator (No Augmentation)
test_datagen = ImageDataGenerator(rescale=1./255)

# Prepare Train and Test Data Generators
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_size, img_size),  # Using img_size for both height and width
    batch_size=batch_size,
    class_mode='categorical'
)

validation_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(img_size, img_size),  # Using img_size for both height and width
    batch_size=batch_size,
    class_mode='categorical'
)

# Define and Compile Model (Example: simple CNN)
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(img_size, img_size, 3)),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(num_classes, activation='softmax')
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the Model
history = model.fit(
    train_generator,
    epochs=50,
    validation_data=validation_generator
)

# Evaluate the Model
loss, accuracy = model.evaluate(validation_generator)
print(f"Validation Accuracy: {accuracy * 100:.2f}%")
```

```python
# Save model architecture to JSON
model_json = model.to_json()
with open("animal_classification_model.json", "w") as json_file:
    json_file.write(model_json)

# Save model weights to H5 file
model.save_weights("animal_classification_model.weights.h5")
print("Model architecture and weights have been saved.")
```
